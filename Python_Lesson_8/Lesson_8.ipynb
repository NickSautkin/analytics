```python
def haversine(fi1, fi2, la1, la2, r = 6371):
    fi1 = np.radians(fi1)
    fi2 = np.radians(fi2)
    la1 = np.radians(la1)
    la2 = np.radians(la2)
    distance = 2 * r * math.asin(((math.sin((fi2 - fi1) / 2))**2 + math.cos(fi1) * math.cos(fi2)*(math.sin((la2 - la1) / 2))**2)**0.5)
    return(distance)
```
