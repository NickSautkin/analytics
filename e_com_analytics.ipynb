Компания начала работать в январе 2015 года и сразу смогла привлечь сотни клиентов в качестве своих пользователей. Пользователи могут быть как на премиальных тарифах (требующих оплаты), так и на бесплатном тарифном плане. Пользователь считается платящим в месяце M, если в этом периоде от него поступила оплата (см файл billing.csv).

### **Датасеты**

**billing.csv** - содержит информацию об оплатах, полученных от клиентов в 2015-2016гг:
* `client_id` - идентификатор клиента
* `country` - страна, в которой находится онлайн-магазин
* `charge_dt` - дата и время (UTC) списания абонентской платы с клиента
* `charge_amount_usd` - сумма, которую списали с клиента


**gtv.csv** - содержит информацию о продажах клиентов (gross transaction volume):
* `period` - период (месяц), в котором были продажи
* `client_id` - идентификатор клиента 
* `country` - страна, в которой находится онлайн-магазин
* `payment_method_id` - метод оплаты, через который клиент (онлайн-магазин) принимал деньги от своих покупателей; у одного клиента может быть несколько методов оплаты
* `gtv_usd` - сумма транзакций, которые были оплачены конечными покупателями клиента (онлайн-магазина) через соответствующий метод оплаты в соответствующем периоде
* `n_orders` - количество заказов, которые были размещены (оплачены) конечными покупателями клиента через соответствующий метод оплаты в соответствующем периоде


Если клиент **впервые** оплатил услуги компании в каком-то месяце, то он будет учтен в двух метриках:
  * **n_FTU** - количество first-time upgraders **за соответствующий месяц**
  * **n_paid** - количество платящих клиентов **по состоянию на конец месяца**

Если клиент платил в прошлом месяце, а в текущем не платил, то на конец текущего месяца он будет учтен в метрике **n_сancelled** (кол-во отменивших подписку клиентов) и не попадет в расчет метрики **n_paid** (по состоянию на конец текущего месяца)

## 3.1

В какое время суток (с точностью до часа) происходит больше всего FTU, основываясь на данных за два года? Постройте гистограмму. Можно считать, что пользователи из каждой страны находятся в столице этой страны.



```python
# датафрейм с датами FTU
task1 = billing.copy()
task1.charge_dt = pd.to_datetime(task1.charge_dt, format='%Y-%m-%d %H:%M:%S')
task1 = task1.groupby(['client_id', 'country'], as_index=False)\
             .agg(dt_FTU=('charge_dt', 'min'))
task1[:3]
```

```python
# словарь используемых часовых поясов
timezones = task1.country.unique()
timezones = pd.DataFrame(timezones).rename(columns={0: 'country'})
timezones['timezone'] = ['Europe/Rome', 'America/Toronto',
                         'Europe/Moscow', 'Australia/Sydney', 'Europe/London']
timezones
```

```python
# добавляем столбец с часовыми поясами в основной df
task1 = task1.merge(timezones, on='country', how='left')

# получаем локальное время и берем из него часы
task1.dt_FTU = task1.apply(lambda x: x['dt_FTU'].tz_localize('utc')
                    .tz_convert(x['timezone']), axis=1)

task1['hour_FTU'] = [i.hour for i in task1.dt_FTU]
task1[:5]
```

```python
task1.hour_FTU.plot.hist(bins=24, grid=False, figsize=(20, 7), fontsize=15)
```

### Чаще всего пользователи совершают оплату в 11 часов дня.

Посчитайте месячный ARPPU по пользователям каждой страны и отобразите на графике (line chart).

```python
# делаем нужный датафрейм из billing
task2 = billing.copy()
task2.charge_dt = pd.to_datetime(task2.charge_dt, format='%Y-%m-%d %H:%M:%S')\
                    .dt.strftime('%Y-%m')
task2[:3]
```

```python
# строим график, но страны надо разделить по столбцам, чтобы каждую показать отдельной кривой.
pd.pivot_table(data=task2, index='charge_dt', columns='country', values='arppu')\
  .plot.line(figsize=(20, 7), xlabel="month", ylabel="ARPPU", fontsize=15)
```

## 3.3

Рассчитайте месячные когорты, показывающие кол-во платящих клиентов на конец каждого периода (retention cohorts). Например, все клиенты, оплатившие в первый раз в марте 2015, попадают в когорту 2015-03 и дальше для месяцев +1, +2, +3, .., +n надо рассчитать количество клиентов из этой когорты, которые были платящими на конец каждого такого периода.

```python
# делаем нужный датафрейм из billing
task3 = billing.copy()
task3 = task3.loc[:, ['client_id', 'charge_dt']]
task3.charge_dt = pd.to_datetime(task3.charge_dt, format='%Y-%m-%d %H:%M:%S')
task3.sample(3)
```

```python
# теперь создаем таблицу с датами первой оплаты клиентом
dates_FTU = task3.groupby('client_id', as_index = False)\
                 .agg(dt_FTU=('charge_dt', 'min'))
dates_FTU.sample(3)
```

```python
# добавляем даты FTU в основной датафрейм и добавляем когорты
task3 = task3.merge(dates_FTU, on='client_id', how='left')
task3['cohort'] = task3.dt_FTU.dt.strftime('%Y-%m')

# считаем столбец с разницой между датами в месяцах
task3['period'] = task3.charge_dt.dt.month - task3.dt_FTU.dt.month + \
                 (task3.charge_dt.dt.year - task3.dt_FTU.dt.year) * 12

# убираем лишние строки
task3 = task3.drop_duplicates(subset=['client_id', 'period'])
task3.sample(3)
```

```python
# аггрегируем данные по пользователям, получаем финальную таблицу.
task3 = task3.groupby(['cohort', 'period'], as_index=False)\
             .agg(clients=('client_id', 'count'))

task3.sample(5)
```

```python
# для удобства можно посмотреть в виде сводной таблицы.
pd.pivot_table(data=task3, index='cohort', columns='period', values='clients').fillna('')
```

### Когорта "2015-01" получилась аномально большой, т.к. в нее попали пользователи из более ранных когорт.
### Также лишние пользователи могли попасть и в другие когорты, если они пропустили оплату в январе 2015го, но платили ранее.
### Поскольку более ранних данных нет, вычленить лишних пользователей не получится.

## 3.4

Используя информацию из п3, рассчитайте для каждой когорты процент пользователей, которые были платными в месяце +1, +2, .., +n по отношению к длине когорты (%-retention relative to cohort length) и отобразите построенную таблицу, используя heatmap.

```python
# чтобы проставить размер когорты в каждой строке, смерджим данные п.3 с ними же после аггрегации
task4 = task3.merge(task3.groupby('cohort', as_index=False)
                         .agg(cohort_size=('clients', 'max'))
                    , on='cohort', how='left')

task4['retention_rate'] = task4.clients / task4.cohort_size

task4[:5]
```

```python
task4_pivot = task4.query("period != 0")\
                   .pivot(index='cohort', columns='period', values='retention_rate')

plt.figure(figsize=(15, 8))
ax = sns.heatmap(task4_pivot, cmap='YlGn', cbar=False, annot=True,
                 fmt='.1%', annot_kws={"fontsize": 10}, linewidths=.5)
ax.set_xlabel('\nMonth number', fontsize=12)
ax.set_ylabel('FTU Cohort\n', fontsize=12)
ax.set_title('Retention rate\n', fontsize=15)
plt.show()
```

## 3.6

Рассчитайте таблицу, содержащую следующие данные помесячно:
  - Количество платящих клиентов на начало месяца (фактически, надо взять кол-во платящих на конец прошлого месяца) (метрика **n_paid_BOP**)
  - Количество новых клиентов, которые начали платить в текущем месяце (это могут быть как FTU, так и те, кто какое-то время не платил, а потом возобновил платежи) (метрика **n_new**)
  - Количество клиентов, которые перестали платить в текущем месяце (**n_cancelled**)
  - Количество платящих клиентов на конец текущего месяца (**n_paid_EOP**)

В итоге, для каждого месяца должно выполняться соотношение:

`n_paid_BOP + n_new - n_cancelled = n_paid_EOP`

```python
# делаем нужный датафрейм из billing
task6_1 = billing.copy()
task6_1.charge_dt = pd.to_datetime(task6_1.charge_dt, format='%Y-%m-%d %H:%M:%S')\
    .to_numpy().astype('datetime64[M]')
task6_1 = task6_1.drop_duplicates(subset=['client_id', 'charge_dt'])

task6_1.sample(3)
```

```python
# делаем финальную таблицу, 2 метрики легко проставить сразу
task6_2 = task6_1.groupby('charge_dt', as_index=False)\
                 .agg(n_paid_EOP=('client_id', 'count'))
task6_2['n_paid_BOP'] = task6_2.n_paid_EOP.shift(1)

task6_2[:5]
```

### Посчитав одну из оставшихся метрик, можно легко посчитать оставшуюся из уравнения, данного в условиях.

### Но надежней будет посчитать обе отдельно, чтобы потом проверить данные на выполнение уравнения.

```python
# функция для расчета n_new
def n_new_func(curr_month):
    cur_m = task6_1.query("charge_dt == @curr_month")

    prev_month = curr_month + pd.DateOffset(months=-1)

    prev_m = task6_1.query("charge_dt == @prev_month")

    n_new = cur_m.merge(prev_m, how='left', on='client_id')\
                 .query("charge_dt_y.isnull()").client_id.count()
    return (n_new)


# функция для расчета n_cancelled
def n_cancelled_func(curr_month):
    cur_m = task6_1.query("charge_dt == @curr_month")

    prev_month = curr_month + pd.DateOffset(months=-1)

    prev_m = task6_1.query("charge_dt == @prev_month")

    n_cancelled_func = cur_m.merge(prev_m, how='right', on='client_id')\
        .query("charge_dt_x.isnull()").client_id.count()

    return (n_cancelled_func)
```

```python
# добавляем оставшиеся метрики
task6_2['n_new'] = task6_2.apply(lambda x: n_new_func(x['charge_dt']), axis=1)
task6_2['n_cancelled'] = task6_2.apply(
    lambda x: n_cancelled_func(x['charge_dt']), axis=1)

# для удобства убираем число из charge_dt
task6_2.charge_dt = task6_2.charge_dt.dt.strftime('%Y-%m')
task6_2 = task6_2.rename(columns={'charge_dt': 'month'})

task6_2
```

### В этой задаче некорректные данные получились только для месяца "2015-01". В зависимости от целей получения этих данных, возможно строку "2015-01" стоит убрать из итоговой таблицы.

```python
# проверяем данные на корректность, первую строку не берем
(task6_2.n_paid_EOP - task6_2.n_paid_BOP - task6_2.n_new + task6_2.n_cancelled)[1:].unique()
```

## 3.8

Посчитайте среднемесячный GTV магазина - общий по всем странам и отдельно для каждой страны (помесячно). Постройте график (line chart).

```python
# делаем нужный датафрейм из gtv
task8 = gtv.copy()
task8 = task8.groupby(['period', 'country', 'client_id'], as_index=False)\
             .agg({'gtv_usd': 'sum'})
task8.sample(3)
```

```python
# датафрейм со среднемесячным GTV по всем странам
task8_a = task8.groupby('period', as_index=False)\
               .agg({'gtv_usd': 'mean'})
task8_a['country'] = 'ALL'

# датафрейм со среднемесячным GTV отдельно по странам
task8_b = task8.groupby(['period', 'country',], as_index=False)\
               .agg({'gtv_usd': 'mean'})

# финальный датафрейм
task8 = pd.concat([task8_a, task8_b], ignore_index=True)
task8.sample(5)
```

```python
sns.set(
    style="whitegrid",
    rc={'figure.figsize': (15, 7)}
)
ax = sns.lineplot(data=task8, x='period', y='gtv_usd',
                  hue='country', linewidth = 2, style='country')
ax.set(xlabel='\nMonth', ylabel='Average GTV\n')
plt.xticks(rotation=35)
plt.show()
```

## 3.9

Посчитайте среднюю сумму заказа - общую по всем странам и отдельно для каждой страны (помесячно). Постройте график (line chart).

```python
# делаем нужный датафрейм из gtv
task9 = gtv.copy()
task9 = task9.groupby(['period', 'country'], as_index=False)\
             .agg({'gtv_usd': 'sum', 'n_orders': 'sum'})
task9['aver_order'] = task9.gtv_usd / task9.n_orders
task9.sample(3)
```

```python
# датафрейм со средними суммами заказа по всем странам
task9_a = task9.groupby('period', as_index=False)\
               .agg({'gtv_usd': 'sum', 'n_orders': 'sum'})
task9_a['country'] = 'ALL'
task9_a['aver_order'] = task9_a.gtv_usd / task9_a.n_orders

# финальный датафрейм
task9 = pd.concat([task9_a, task9], ignore_index=True)
task9.sample(5)
```

```python
sns.set(
    style="whitegrid",
    rc={'figure.figsize': (15, 7)}
)
ax = sns.lineplot(data=task9, x='period', y='aver_order',
                  hue='country', linewidth = 2, style='country')
ax.set(xlabel='\nMonth', ylabel='Average Order Sum\n')
plt.xticks(rotation=35)
plt.show()
```
