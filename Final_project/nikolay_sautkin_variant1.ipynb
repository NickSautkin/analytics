Файлы содержат данные за очень длинный период - 22 года.
    
Для удобства можно указывать начало и окончание интересуемого периода регистраций в качестве аргументов функции.
    
Также аргументом сделаем период дней, за который нужно посмотреть retention.
    
Приступим к написанию функции. 


```python
def retention(registrations_path, authorisations_path, reg_start, reg_end, retention_period):
    '''
    Возвращает данные о retention в виде heatmap.
            Параметры:
                    registrations_path (str): путь к файлу с регистрациями
                    authorisations_path (str): путь к файлу с авторизациями
                    reg_start (str): первая дата интересуемого периода регистраций
                    reg_end (str): последняя дата интересуемого периода регистраций
                    retention_period (int): кол-во дней для retention
            Возвращаемое значение:
                    retention_heatmap : heatmap
    '''
    # преобразуем введенные даты в корректный формат для дальнейших операций
    reg_start = pd.to_datetime(reg_start).date()
    reg_end = pd.to_datetime(reg_end).date()

    # подтягиваем по указанному пути данные о регистрациях,
    # преобразовываем столбец с датами и убираем данные, не входящие в интересующий нас период
    registrations = pd.read_csv(registrations_path, sep=';')
    registrations.reg_ts = pd.to_datetime(registrations.reg_ts, unit='s')\
                             .dt.date
    registrations = registrations.query(
        "reg_ts >= @reg_start and reg_ts <= @reg_end")

    # аналогично для данных по авторизациям
    authorisations = pd.read_csv(authorisations_path, sep=';')
    authorisations.auth_ts = pd.to_datetime(authorisations.auth_ts, unit='s')\
                               .dt.date
    authorisations = authorisations.query("auth_ts >= @reg_start")

    # объединяем данные в один dataframe,
    # создаем столбец c кол-вом дней с регистрации до текущей авторизации,
    # и меняем ему формат на более удобный числовой
    total_df = registrations.merge(authorisations, on='uid', how='left')
    total_df['day_number'] = total_df.auth_ts - total_df.reg_ts
    total_df.day_number = total_df.day_number.dt.days

    # убираем из данных авторизации, совершенные позже интересующего нас периода.
    # группируем данные и считаем кол-во уникальных пользователей по дням
    total_df = total_df.query("day_number <= @retention_period")\
                       .groupby(['reg_ts', 'day_number'], as_index=False)\
                       .agg({'uid': 'nunique'})\

    # для подсчета retention rate нам нужен столбец с кол-вом регистраций в текущий день.
    # для его создание мерджим datafreme с его же сгруппированной версией.
    # и сразу для удобства переименовываем колонки и убираем данные нулевого дня, где rr всегда 100%
    total_df = total_df.merge(total_df.groupby('reg_ts', as_index=False)
                                      .agg({'uid': 'max'}),
                              on='reg_ts', how='left')\
                       .rename(columns={'reg_ts': 'registration_date', 'uid_x': 'users_auths', 'uid_y': 'users_regs'})\
                       .query('day_number > 0')

    # теперь можно создать столбец с retention rate и сформировать финальную таблицу
    total_df['retention_rate'] = (total_df.users_auths / total_df.users_regs)
    total_df = total_df.pivot(
        index='registration_date', columns='day_number', values='retention_rate')

    # настраиваем и выводим heatmap
    plt.figure(figsize=(15, 8))
    ax = sns.heatmap(total_df, cmap='YlGn', cbar=False,
                     annot=True, fmt='.1%', annot_kws={"fontsize": 10}, linewidths=.5)
    ax.set_xlabel('\nDay number')
    ax.set_ylabel('Registration date\n')
    ax.set_title('Retention rate\n', fontsize=15)
    plt.show()
```

```python
# проверяем работу функции
retention('problem1-reg_data.csv', 'problem1-auth_data.csv',
          '2020-09-01', '2020-09-20', 7)
```

---

<div style="border:solid LightGrey 4px; padding: 15px"> 

На случай, если группы все же сформированы корректно, проверим, если ли статистически значимые различия в их результатах.

Думаю, нет смысла проверять их распределения на нормальность - по гистограммам уже видно, что они ненормальны.
Но большой размер выборок позволяет нам использовать T-тест.
    
Начнем со сравнения revenue, из которой также вытекает ARPU.
    
    H0 - различий по revenue у групп нет
    H1 - группы различаются по revenue

```python
pg.ttest(x=df_ab_results[df_ab_results.testgroup == 'a'].revenue,
         y=df_ab_results[df_ab_results.testgroup == 'b'].revenue)
```

<div style="border:solid LightGrey 4px; padding: 15px"> 
Результат говорит, что вероятность случайно получить текущие или бОльшие различия по revenue составляет 0.53, что очень много.
    
Поэтому мы принимаем нулевую гипотезу, различий по revenue у групп нет.

---

<div style="border:solid LightGrey 4px; padding: 15px"> 
Также проверим на стат.значимость различие в conversion. 
Для этого добавим в dataframe дополнительный столбец с указанием, совершил пользователь покупку или нет.

И после этого проведем T-test:
    
    H0 - различий по revenue у групп нет
    H1 - группы различаются по revenue

```python
df_ab_results['purchased'] = df_ab_results.revenue > 0
df_ab_results[:3]
```

```python
pg.ttest(x=df_ab_results[df_ab_results.testgroup == 'a'].purchased,
         y=df_ab_results[df_ab_results.testgroup == 'b'].purchased)
```

<div style="border:solid LightGrey 4px; padding: 15px"> 
Этот тест дал достаточно маленький p-value, мы можем отклонить нулевую гипотезу - различия по конверсии у групп стат.занчимые.

Если руководствоваться только результатами тестов, то можно сделать вывод, что тестовая группа показала себя хуже по показателю конверсии в покупки, и не отличается от контрольной по доходу с охваченного акцией пользователя.
